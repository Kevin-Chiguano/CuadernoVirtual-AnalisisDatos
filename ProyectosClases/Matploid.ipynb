```python
import pandas as pd

# Cargar el archivo titanic.csv
df = pd.read_csv('titanic.csv')

# Añadir una nueva columna booleana 'Menor_de_edad' (True si 'Age' < 18)
df['Menor_de_edad'] = df['Age'] < 18

# Mostrar las primeras filas para verificar
print(df)
```

```python
# Calcular los porcentajes de supervivencia por clase y si eran menores de edad o no
survival_rate = df.groupby(['Pclass', 'Menor_de_edad'])['Survived'].mean() * 100

# Mostrar los resultados
print(survival_rate)
```

```python
import matplotlib.pyplot as plt

# Configurar los datos para los histogramas
classes = df['Pclass'].unique()
menores = survival_rate.loc[:, True]
mayores = survival_rate.loc[:, False]

# Dibujar histogramas
fig, ax = plt.subplots(1, 2, figsize=(12, 6), sharey=True)

# Histograma para menores de edad
ax[0].bar(classes, menores, color='blue')
ax[0].set_title('Porcentaje de menores de edad que sobrevivieron')
ax[0].set_xlabel('Clase')
ax[0].set_ylabel('Porcentaje')

# Histograma para mayores de edad
ax[1].bar(classes, mayores, color='green')
ax[1].set_title('Porcentaje de mayores de edad que sobrevivieron')
ax[1].set_xlabel('Clase')

# Mostrar los gráficos
plt.tight_layout()
plt.show()
```
